# Webscrapping Using BeautifulSoup

```python
#import libraries needed
import pandas as pd
import requests
from bs4 import BeautifulSoup

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Here we will webscrap Unicorn companies using BeautifulSoup.

We will use the website [https://www.cbinsights.com/research-unicorn-companies](https://www.cbinsights.com/research-unicorn-companies) for this notebook.

```python
url = "https://www.cbinsights.com/research-unicorn-companies"
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data, 'html5lib')
```

```python
unicorn_data = pd.DataFrame(columns=["Company", "Valuation", "Date_Joined", "Country", "City", "Industry", "Select_Investors"])

# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    company = col[0].text
    valuation = col[1].text
    joined = col[2].text
    country = col[3].text
    city = col[4].text
    industry = col[5].text
    investors = col[6].text
    
    # Finally we append the data of each row to the table
    unicorn_data = unicorn_data.append({"Company":company, "Valuation":valuation, "Date_Joined":joined, "Country":country, "City":city, "Industry":industry, "Select_Investors":investors}, ignore_index=True)  
```

```python
#view webscrap data
unicorn_data
```

```python
#save as csv file
unicorn_data.to_csv('unicorn_data_web_scraped.csv', index=False)
```
